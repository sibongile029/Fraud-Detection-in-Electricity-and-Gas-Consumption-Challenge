```r
C <- invoice_train$counter_type
elec <- (sum(C == "ELEC") / length(C)) * 100
gaz <- (sum(C == "GAZ") / length(C)) * 100

# Create the pie chart
par(mar = c(1, 1, 4, 1))  # Adjust the margin to make space for the border
pie(c(elec, gaz), labels = c("ELEC", "GAZ"), col = c("#00eaff", "#bbff00"),
    main = "Proportion of Counter type (ELEC to GAZ)",
    border = "white")

# Add percentage labels
percent_labels <- paste0(round(c(elec, gaz), 1), "%")
legend("topright", legend = c("ELEC", "GAZ"), fill = c("#00eaff", "#bbff00"))
legend("bottomright", legend = percent_labels, cex = 0.8, bty = "n")

```
